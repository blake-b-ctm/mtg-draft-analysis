```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/game_data_public.LTR.TradSealed.csv')
df.head()
```

```python
pd.set_option('display.max_columns', 18)
df.iloc[:, : 18]
```

```python
    # cty_df = census_df[census_df['SUMLEV']==50]
    # largest_counties = cty_df.groupby('STNAME')['CENSUS2010POP'].nlargest(3)
    # sum_of_largest_counties = largest_counties.groupby('STNAME').sum()
sample_size = len(df) / 200
print(sample_size, 'sample size')
main_colors = df.groupby('main_colors')['main_colors'].count().loc[lambda x : x >= sample_size]
main_colors
```

```python
wins_per_color_pair = df.groupby('main_colors')['won'].sum()
wins_per_color_pair
```

```python
winrate_per_color = (wins_per_color_pair / main_colors).dropna().map("{:.2%}".format)
# winrate_per_color.sort_values(ascending=False)

color_wins = pd.concat([main_colors, winrate_per_color], axis=1)
color_wins.columns = ['games_played', 'win_rate']
color_wins.sort_values(by='win_rate', ascending=False)
```
